**Action plan:**

In project plan is divide into few parts that are as follows:

 **Understanding the problem:** 

The depth understanding of this topic is must for the success of this project. Evaluation will be done in the first phase of the project to decide what kind of attacks are major threat. For this task, specific set of forensics experts and security researchers will be allowed to work on it. 

**Extraction of data:**

Second phase of the project will start that is extracting specific parameters from the collected information by analysts. 
The data will be gathered from various sources such as feedback, surveys, questionaries, data provided by government and data that is recorded by organization during the ongoing attack on their IT infrastructure. The collected data will contain various attack parameters, vulnerabilities that were exploited by the attacker, type of virus or exploit used to gain access to system. Also, the motive behind the attack will be recorded such as what was the intention of attacker to steal data or to infect the server. 

**Developing the model:**

To develop the model, it is required to convert the extracted data into useful and understandable form of data. The main thing is to choose useful attributes from the extracted data and perform data classification and data pre-processing. For data classification, classifiers like Random Forest, Linear Regression or Support Vector Machine can be used. After that training of models will be done. 

**Testing the prototype:**

After developing the successful model, the next step will be to develop a small prototype and test this prototype in a small controlled environment to test and evaluate the results. In this controlled environment specific conditions will be created regarding different cyber-security attacks and this model will be tested against these attacks. The outcomes will be in the form that cyber attack will happen or no.
The results and this small prototype will be sent to specific organizations with a trial license of small periods and the performance of this prototype will be recorded using a remote server and later on the results will be matched with the pre-tests that were performed inside the controlled environment and the defects and good points will be taken into consideration to produce the final working model.

**Final Application:**

After careful evaluation of the prototype the final model will be generated. The cost and timeline for the final model will be generated. The prototype will be created on a large-scale model.
