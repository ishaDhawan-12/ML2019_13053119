Fisherface:
Fisherface method is developed based on Linear Discriminant Analysis. It focuses on maximizing the inter-class difference. It has no role in dealing with data representation. In other words, determinant of a between-class matrix is increased, and determinant of within-class matrix is decreased. Both these tasks are done simultaneously. In Fisherfaces, it was found that linear projections can be used to deduce the dimensions as classes are convex.
Eigenface:
The base of Eigenface implementation is PCA (Principle Computer Analysis). The goal of this method is to decrease the dimensionality of an image matrix. A linear transformation is said to be used by this method for face recognition in illumination or pose change.
There are two kinds of datasets in the paper, one is the test, and other one is learning. The test set contains unlabelled pictures and learning set contain labeled pictures with the identity of human beings.
